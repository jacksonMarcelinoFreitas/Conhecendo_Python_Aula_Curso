```python
'''Faça um programa que solicita o nome e a nota de vários alunos.
O programa deve perguntar, a cada iteração, se deseja continuar o cadastro.
Ao final devem ser apresentados todos os alunos e  notas  cadastrados,  a  média
da  turma,  lista  dos  alunos  aprovados  e  a  lista  dos  alunos reprovados
(média  de  aprovação  >=  6.0).  Utilize  listas  para:  alunos,  notas,  aprovados
e reprovados.'''

continuar = 's'
nome = []
nota = []

while continuar.lower() == 's':
    nome.append(input("\nInforme o nome: "))
    nota.append(int(input("Informe a nota")))
    continuar = input("\nDeseja continuar ?\n\n")

print("\nNomes e notas CADASTRADOS\n ")
for i in range(len(nome)):
    print("%s -> %s "%(nome[i], nota[i]))

print("\nNomes e notas dos aluno REPROVADOS\n ")
for i in range(len(nome)):
    if(nota[i] < 6):
        print("Aluno: %s Nota: %d " %(nome[i], nota[i]))


print("Nomes e notas dos aluno APROVADOS\n ")
for i in range(len(nome)):
    if(nota[i] >= 6):
        print("Aluno: %s Nota: %d " %(nome[i], nota[i]))


somaNotas = sum(nota)
mediaSala = somaNotas/len(nome)
print("A soma das notas é %d" %(somaNotas))
print("A media da turma  é  %d" %(mediaSala))


        
    
    
    
```
